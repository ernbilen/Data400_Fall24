```python
from selenium import webdriver
import pandas as pd
import time
import random

driver = webdriver.Chrome('/Users/eren/Downloads/chromedriver')

# scrape a single page

driver.get('https://forums.edmunds.com/discussion/2864/general')

def get_comment_ids():
    ids = driver.find_elements('xpath','//*[contains(@id, "Comment_")]')
    comment_ids = []
    for i in ids:
        comment_ids.append(i.get_attribute('id'))
    return comment_ids


comment_ids = get_comment_ids()

comments=[]
for i in comment_ids:
    userid_element = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[2]/div[1]/span[1]')
    userid = userid_element.text

    user_date = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[2]/div[2]/span/a/time')
    date = user_date.get_attribute('title')

    user_comment = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[3]/div')
    comment = user_comment.text
    
    comments.append([userid,date,comment])
    
    
df = pd.DataFrame(comments,columns=['userid','date','comment'])
df
```

```python
# multi-page option 1: loop over link
from selenium import webdriver
import pandas as pd
import time
import random

driver = webdriver.Chrome('/Users/eren/Downloads/chromedriver')

def get_comment_ids():
    ids = driver.find_elements('xpath','//*[contains(@id, "Comment_")]')
    comment_ids = []
    for i in ids:
        comment_ids.append(i.get_attribute('id'))
    return comment_ids

# scrape a multi page, option 1

mylist = ['p1','p2','p3']

comments=[]
for j in mylist:

    driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/'+j)
    comment_ids = get_comment_ids()
    for i in comment_ids:
        userid_element = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[2]/div[1]/span[1]')
        userid = userid_element.text

        user_date = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[2]/div[2]/span/a/time')
        date = user_date.get_attribute('title')

        user_comment = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[3]/div')
        comment = user_comment.text

        comments.append([userid,date,comment])
    
df = pd.DataFrame(comments,columns=['userid','date','comment'])
df
```

```python
# multi-page scraping option 2: click on "next page" button

from selenium import webdriver
import pandas as pd
import time
import random

driver = webdriver.Chrome('/Users/eren/Downloads/chromedriver')

def get_comment_ids():
    ids = driver.find_elements('xpath','//*[contains(@id, "Comment_")]')
    comment_ids = []
    for i in ids:
        comment_ids.append(i.get_attribute('id'))
    return comment_ids

def get_single_page():
    comment_ids = get_comment_ids()
    for i in comment_ids:
        userid_element = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[2]/div[1]/span[1]')
        userid = userid_element.text

        user_date = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[2]/div[2]/span/a/time')
        date = user_date.get_attribute('title')

        user_comment = driver.find_element('xpath','//*[@id="' + i + '"]/div/div[3]/div')
        comment = user_comment.text

        comments.append([userid,date,comment])
    return comments
    

# scrape a multi page, option 2
comments=[]
driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/p1')

for j in range(3):
    comments = get_single_page()
    button = driver.find_element('xpath','//*[@title="Next Page"]')
    button.click()
    time.sleep(random.uniform(2,3)) # sleep for a random amount of time between e.g., 2 and 3 seconds
    
df = pd.DataFrame(comments,columns=['userid','date','comment'])
df
```
